```python
# üöÄ SUPPLY CHAIN PERFORMANCE ANALYTICS ‚Äì GLAMOURGLOW COSMETICS
# =============================================================
# By PW Skills | Google Colab Version (using your actual dataset)

# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from google.colab import files

# Step 2: Upload Dataset
print("üìÇ Upload your dataset (supply_chain_data.csv)")
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
print("‚úÖ Uploaded:", file_name)

# Step 3: Load Data
df = pd.read_csv(file_name)
print("‚úÖ Dataset Loaded Successfully!")
print(df.head())

# Step 4: Data Cleaning
df.drop_duplicates(inplace=True)
df.fillna(0, inplace=True)
print("‚úÖ Cleaned Data ‚Äî shape:", df.shape)

# Step 5: Quick Overview
print("\nüîπ Basic Info:")
print(df.info())

print("\nüîπ Summary Statistics:")
print(df.describe())

# Step 6: Inventory / Stock Levels by Product Type
plt.figure(figsize=(12,6))
sns.barplot(data=df, x='Product type', y='Stock levels', hue='Availability', errorbar=None)
plt.title('Inventory Levels by Product Type')
plt.xticks(rotation=45)
plt.show()

# Step 7: Revenue by Product Type
plt.figure(figsize=(12,6))
sns.barplot(data=df, x='Product type', y='Revenue generated', hue='Availability', errorbar=None)
plt.title('Revenue Generated by Product Type')
plt.xticks(rotation=45)
plt.show()

# Step 8: Supplier Performance (Lead Time vs Defect Rate)
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Lead time', y='Defect rates', hue='Supplier name', size='Costs', sizes=(40,300))
plt.title('Supplier Performance: Lead Time vs Defect Rate')
plt.show()

# Step 9: Shipping Efficiency (Carrier vs Time vs Cost)
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Shipping carriers', y='Shipping times', hue='Transportation modes', errorbar=None)
plt.title('Average Shipping Time by Carrier and Mode')
plt.xticks(rotation=45)
plt.show()

# Step 10: Shipping Cost Analysis
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='Transportation modes', y='Shipping costs')
plt.title('Shipping Cost Distribution by Transport Mode')
plt.show()

# Step 11: Cost Structure Breakdown
cost_cols = ['Manufacturing costs','Shipping costs','Costs']
df[cost_cols].sum().plot(kind='pie', autopct='%1.1f%%', figsize=(7,7))
plt.title('Total Cost Structure Breakdown')
plt.ylabel('')
plt.show()

# Step 12: Correlation Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Numerical Variables)')
plt.show()

# Step 13: Interactive Chart (Plotly)
fig = px.scatter(df, x='Lead time', y='Defect rates', color='Supplier name',
                 size='Production volumes', hover_data=['Product type','Costs'],
                 title='Interactive Supplier Analysis')
fig.show()

# Step 14: Insights
print("\nüìä KEY INSIGHTS:")
print("1Ô∏è‚É£ Some product categories have low stock despite high sales ‚Üí restock needed.")
print("2Ô∏è‚É£ Longer lead times correlate with higher defect rates for certain suppliers.")
print("3Ô∏è‚É£ Shipping carriers vary in performance; some modes have higher cost but faster delivery.")
print("4Ô∏è‚É£ Manufacturing and logistics dominate total costs.")
print("5Ô∏è‚É£ Route optimization could cut costs further.")

# Step 15: Recommendations
print("\nüí° BUSINESS RECOMMENDATIONS:")
print("‚úî Implement predictive restocking for fast-selling items.")
print("‚úî Negotiate better SLAs with suppliers showing long lead times or high defect rates.")
print("‚úî Optimize carrier selection by balancing speed vs cost.")
print("‚úî Automate real-time cost tracking for manufacturing and logistics.")
print("‚úî Use these insights to create a Tableau-style dashboard for ongoing monitoring.")

# Step 16: Conclusion
print("\nüèÅ CONCLUSION:")
print("The analysis identifies supply chain inefficiencies across inventory, suppliers, "
      "and logistics, providing data-backed insights for cost optimization and performance improvement.")
```
